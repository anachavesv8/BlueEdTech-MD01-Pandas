```python
import pandas as pd
import math

df = pd.read_csv('/content/Musicas.csv')
df
```

```python
grupo = df.sample(3).reset_index()
d1 = grupo.loc[0, 'danceability']
e1 = grupo.loc[0, 'energy']
grupo1 = [d1, e1]

d2 = grupo.loc[1, 'danceability']
e2 = grupo.loc[1, 'energy']
grupo2 = [d2, e2]

d3 = grupo.loc[2, 'danceability']
e3 = grupo.loc[2, 'energy']
grupo3 = [d3, e3]

def dist_euclidiana(d1, d2, e1, e2):
    return math.sqrt(((d2-d1)**2) + ((e2 - e2)**2))

def mapear_grupo(x):
    dist1 = dist_euclidiana(grupo1[0], x['danceability'], 
                            grupo1[1], x['energy'])
    dist2 = dist_euclidiana(grupo2[0], x['danceability'], 
                            grupo2[1], x['energy'])
    dist3 = dist_euclidiana(grupo3[0], x['danceability'], 
                            grupo3[1], x['energy'])
    if dist1 < dist2 and dist3:
        return 'grupo1'
    elif dist2 < dist3:
        return 'grupo2'
    else:
        return 'grupo3'
```

```python
import plotly.express as px
```

```python
df['grupo'] = df.apply(mapear_grupo, axis=1)

df.head()
```

```python
fig = px.scatter(df, x='energy', 
                     y='danceability',
                     color='grupo')
```
