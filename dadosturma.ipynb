```python
def signos(x):
  if (x['day'] >= 21 and x['month'] == 4) or (x['day'] <= 20 and x['month'] == 5):
    return 'Touro'
  if (x['day'] >= 21 and x['month'] == 5) or (x['day'] <= 20 and x['month'] == 6):
    return 'Gemeos'
  if (x['day'] >= 21 and x['month'] == 6) or (x['day'] <= 22 and x['month'] == 7):
    return 'Cancer'
  if (x['day'] >= 23 and x['month'] == 7) or (x['day'] <= 22 and x['month'] == 8):
    return 'Leão'
  if (x['day'] >= 23 and x['month'] == 8) or (x['day'] <= 22 and x['month'] == 9):
    return 'Virgem'
  if (x['day'] >= 23 and x['month'] == 9) or (x['day'] <= 22 and x['month'] == 10):
    return 'Libra'
  if (x['day'] >= 23 and x['month'] == 10) or (x['day'] <= 21 and x['month'] == 11):
    return 'Escorpião'
  if (x['day'] >= 22 and x['month'] == 11) or (x['day'] <= 21 and x['month'] == 12):
    return 'Sagitario'
  if (x['day'] >= 22 and x['month'] == 12) or (x['day'] <= 19 and x['month'] == 1):
    return 'Capricornio'
  if (x['day'] >= 20 and x['month'] == 1) or (x['day'] <= 18 and x['month'] == 2):
    return 'Aquario'
  if (x['day'] >= 19 and x['month'] == 2) or (x['day'] <= 20 and x['month'] == 3):
    return 'Peixes'
  else:
    return 'Aries'
```

```python
df['signos'] = df.apply(signos,axis = 1)
df
```
