*Classificação das especies de flores:*

Utilizada na definição de função abaixo.

1.   especie = 0: Setosa
2.   especie = 1: Versicolor
3.   especie = 2: Virginica


```python
def arvore1(petal_length, petal_width, sepal_width, especie):
  if petal_length <= 2.45:
    especie = 0
    return especie
  elif petal_width <= 1.75:
    if petal_length <=4.95:
      if petal_width <= 1.65:
        especie = 1
        return especie
      else:
        especie = 2
        return especie
    elif petal_width <=1.55:
      especie = 2
      return especie
    else:
      especie = 1
      return especie
  elif petal_length <= 4.85:
    if sepal_width <= 3.1:
      especie = 2
      return especie
    else:
      especie = 1
      return especie
  else:
    especie = 2
    return especie
```

# **Avaliação de modelo**

*Implementar a métrica para avaliar os modelos (1,5 ponto)*

*Definir o melhor modelo e justificar (1 ponto)*

*Modelo Arvore 1*

```python
#Ajustando a coluna Arvore2 para float64
df1.info()
```

```python
df1[['Arvore1','Arvore2']] = df1[['Arvore1','Arvore2']].astype(float, errors = 'raise')
df1.info()
```

```python
#Definindo função para avaliar os erros do modelo da arvore1.
def avaliacao_arvore1(df1):
  if df1['target'] == df1['Arvore1']:
    return 1
  else:
    return 0
```

```python
df1['Avaliacao_arvore1'] = df1.apply(avaliacao_arvore1,axis = 1)
```

```python
df1[['target','Arvore1','Avaliacao_arvore1']]
```

```python
df1['Avaliacao_arvore1'].sum()/150
```

O modelo de classificação da Arvore 1, Teve uma porcentagem de acerto de 99%.

*Modelo Arvore 2*

```python
#Definindo função para avaliar os erros do modelo da arvore2.
def avaliacao_arvore2(df1):
  if df1['target'] == df1['Arvore2']:
    return 1
  else:
    return 0
```

```python
df1['Avaliacao_arvore2'] = df1.apply(avaliacao_arvore2,axis = 1)
```

```python
df1[['target','Arvore2','Avaliacao_arvore2']]
```

```python
df1['Avaliacao_arvore2'].sum()/150
```

O modelo de classificação da Arvore 2, Teve uma porcentagem de acerto de 96%.
