```python
import pandas as pd
dados_gov = pd.read_csv('/content/1-dadosgovbr---2014.txt', sep=';', encoding='latin')
```

```python
dados_gov.info()
```

```python
dados_gov.head()
```

Qual a quantidade de reclamações registradas? 

```python
dados_gov['Avaliação Reclamação']
```

```python
def reclamações(dados_gov):
  if dados_gov['Avaliação Reclamação'] == 'Não Resolvida':
    reclamações = 0
    return reclamações
  elif dados_gov['Avaliação Reclamação'] == 'Não Avaliada':
    reclamações = 1
    return reclamações
  elif dados_gov['Avaliação Reclamação'] == 'Resolvida':
    reclamações = 2
    return reclamações
  else:
    reclamações = 3
    return reclamações
```

```python
dados_gov['Avaliações'] = dados_gov.apply(reclamações, axis = 1)
dados_gov['Avaliações']
```

```python
filtro = dados_gov[dados_gov['Avaliações'] == 0] #11200 reclamações Não resolvidas
filtro.describe()
```

```python
filtro2 = dados_gov[dados_gov['Avaliações'] == 1] #13526 reclamações Não avaliadas
filtro2.describe()
```

```python
filtro3 = dados_gov[dados_gov['Avaliações'] == 2] #17427 reclamações resolvidas
filtro3.describe()
```

```python
filtro4 = dados_gov[dados_gov['Avaliações'] == 3] #1834 NAN
filtro4.describe()
```

Qual é o tempo médio, máximo e mínimo de resposta?

```python
dados_gov.agg({"Tempo Resposta" : ['min','max','mean']})
```

Qual é a nota média, máxima e mínima do consumidor? 

```python
dados_gov.agg({'Nota do Consumidor' : ['min','max','mean']})
```

Qual a quantidade de reclamações por Sexo?

```python
dados_gov['Sexo'].value_counts()
```

Qual a quantidade de reclamações por Estado?

```python
dados_gov['UF'].value_counts()
```

Qual é a porcentagem de reclamações registradas e não respondidas?


```python
dados_gov['Avaliação Reclamação'].value_counts(normalize=True)
```
